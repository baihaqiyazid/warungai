```python
import datetime
from typing import List

def conversational_history(history: list, current_message: str) -> List:
    """
    Memformat riwayat percakapan dan pesan terbaru agar mudah dipahami agen.

    Args:
        history (list): Daftar riwayat percakapan [timestamp, role, message].
        current_message (str): Pesan terbaru dari pengguna.

    Returns:
        str: String riwayat percakapan yang diformat untuk konteks agen.
    """
    conversation_history_for_agent = []

    # Format riwayat percakapan yang sudah ada
    for entry in history:
        timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        user_msg, assistant_msg = entry
        conversation_history_for_agent.append(f"{timestamp} | buyer: {user_msg}")
        conversation_history_for_agent.append(f"{timestamp} | assistant: {assistant_msg}")

    # Tambahkan pesan baru dari pembeli ke riwayat
    current_timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    conversation_history_for_agent.append(f"{current_timestamp} | buyer: {current_message}")
    
    # Gabungkan riwayat menjadi string
    formatted_history_string = "\n".join(conversation_history_for_agent)

    full_request = f"""
Konteks Percakapan Sebelumnya::
---
{formatted_history_string}
---

--- Percakapan Berlanjut ---
Pesan terbaru dari pembeli: {current_message}

"""
    return conversation_history_for_agent, full_request
```
