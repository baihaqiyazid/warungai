```python
import gradio as gr
from pyzbar.pyzbar import decode
import cv2
import numpy as np
import json

def process_uploaded_image(image):
    unique_ids = set()
    if image is None:
        return "No image received."

    # Convert PIL image (from Gradio) to OpenCV format
    frame = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
    qrcodes = decode(frame)

    for qrcode in qrcodes:
        try:
            data = qrcode.data.decode('utf-8')
            obj = json.loads(data)
            qr_id = obj.get("id")
            if qr_id:
                unique_ids.add(str(qr_id))
        except:
            continue

    if unique_ids:
        return list(unique_ids)
    return "No QR ID found."

# JavaScript untuk mengotomatiskan capture setiap 5 detik
custom_js = """
() => {
    setInterval(() => {
        const captureButton = document.querySelector('button[aria-label="capture photo"]');
        const submitButton = document.querySelector('button.lg.primary.svelte-1ixn6qd');
        
        if (captureButton) {
            captureButton.click();
            // Tunggu sebentar untuk memastikan capture selesai, lalu klik submit
            setTimeout(() => {
                if (submitButton) {
                    submitButton.click();
                }
            }, 500); // Delay 500ms untuk memastikan capture selesai
        }
    }, 5000); // 5000 ms = 5 detik
}
"""

gr.Interface(
    fn=process_uploaded_image,
    inputs=gr.Image(type="pil", label="Scan QR dari Kamera atau Upload Gambar", sources="webcam"),
    outputs="json",
    title="Scan QR Produk",
    description="Tunjukkan QR produk ke kamera. Sistem akan otomatis mengambil gambar setiap 5 detik dan menampilkan ID-nya.",
    js=custom_js  # Tambahkan JavaScript kustom
).launch()
```

## Custom Hook

```python
from typing import Any
from agents import Tool, FunctionTool, RunContextWrapper, RunHooks, Usage
from agents.agent import Agent


class CustomHooks(RunHooks):

    def __init__(self):
        self.event_counter = 0
    
    def _usage_to_str(self, usage: Usage): 
        """Convert Statistics to a string"""
        return (f"{usage.requests} request, {usage.input_tokens} input tokens", 
                f"{usage.output_tokens} output_tokens, {usage.total_tokens} input total_tokens", 
                )

    async def on_start(self, context: RunContextWrapper, agent: Agent) -> None:
        print(f"AGENT {agent.name} START") 

    async def on_end(self, context: RunContextWrapper, agent: Agent, output: Any) -> None:
        print(f"AGENT {agent.name} END")

    async def on_tool_start(self, context: RunContextWrapper, agent: Agent, tool: Tool) -> None:
        print(f"Tool {tool.name} START")

    async def on_tool_end(self, context: RunContextWrapper, agent: Agent, tool: Tool, config) -> None:
        print(f"Tool {tool.name} END")
       

custom_hook = CustomHooks()
    
```

```python
from agents import FunctionToolResult, ToolsToFinalOutputResult, RunContextWrapper
import pandas as pd

current_transaction_details_df = pd.DataFrame(columns=["Produk", "Jumlah", "Harga per Produk", "Total Harga Produk"]) # Initialized global DataFrame

def my_tools_to_final_output_function(
    run_context: RunContextWrapper,
    tool_results: List[FunctionToolResult]
) -> ToolsToFinalOutputResult:
    
    global current_transaction_details_df # Declare global

    for i in tool_results:
        if i.tool.name == 'create_transaction' or i.tool.name == 'update_transaction':
            json_obj = json.loads(i.output)
            print(json_obj)
            v = json.loads(json_obj['text'])
            print(v)

            details = []
            for detail in v['detail_transactions']:
                produk_name = "Unknown Product"
                # Find product name from the global products list using produk_id
                for p in products: # 'products' is a global list loaded earlier in the notebook
                    if p.get('id') == detail.get('produk_id'):
                        produk_name = p.get('nama_barang', "Unknown Product")
                        break
                details.append({
                    "Produk": produk_name,
                    "Jumlah": detail["qty"],
                    "Harga per Produk": detail["harga_per_produk"],
                    "Total Harga Produk": detail["total_harga_produk"]
                })
            current_transaction_details_df = pd.DataFrame(details)

            if v['transaction']['status'] == 'success':
                current_transaction_details_df = pd.DataFrame(columns=["Produk", "Jumlah", "Harga per Produk", "Total Harga Produk"]) # Initialized global DataFrame

            print(v['transaction'])
            print(v['detail_transactions'])

    combined_output = "\n".join([result.output for result in tool_results])

    return ToolsToFinalOutputResult(
        final_output=combined_output,
        is_final_output=False  
    )
```

```python
import datetime
from typing import List

def conversational_history(history: list, current_message: str) -> List:
    """
    Memformat riwayat percakapan dan pesan terbaru agar mudah dipahami agen.

    Args:
        history (list): Daftar riwayat percakapan [timestamp, role, message].
        current_message (str): Pesan terbaru dari pengguna.

    Returns:
        str: String riwayat percakapan yang diformat untuk konteks agen.
    """
    conversation_history_for_agent = []

    # Format riwayat percakapan yang sudah ada
    for entry in history:
        timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        user_msg, assistant_msg = entry
        conversation_history_for_agent.append(f"{timestamp} | buyer: {user_msg}")
        conversation_history_for_agent.append(f"{timestamp} | assistant: {assistant_msg}")

    # Tambahkan pesan baru dari pembeli ke riwayat
    current_timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    conversation_history_for_agent.append(f"{current_timestamp} | buyer: {current_message}")
    
    # Gabungkan riwayat menjadi string
    formatted_history_string = "\n".join(conversation_history_for_agent)

    full_request = f"""
Konteks Percakapan Sebelumnya::
---
{formatted_history_string}
---

--- Percakapan Berlanjut ---
Pesan terbaru dari pembeli: {current_message}

"""
    return conversation_history_for_agent, full_request
```

```python
from collections import defaultdict
from produk import ProdukService

def get_produk_by_image(unique_ids):
    produk_counter = defaultdict(lambda: {"nama": "", "jumlah": 0})

    for id_str in unique_ids:
        produk = ProdukService.get_produk(int(id_str))
        if produk:
            produk_id = produk.id
            produk_counter[produk_id]["nama"] = produk.nama_barang
            produk_counter[produk_id]["jumlah"] += 1

    hasil_ringkasan = ["Saya mau beli barang dibawah ini:"]
    for produk_id, data in produk_counter.items():
        hasil_ringkasan.append(
            f"product_id={produk_id} {data['nama']} dengan jumlah {data['jumlah']}"
        )

    gabungan = "\n".join(hasil_ringkasan)
    return gabungan
    

# Contoh pemanggilan
get_produk_by_image(['1', '2', '2', '2', '3'])
```
