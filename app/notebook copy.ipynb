```python
import gradio as gr
import os
from setup_logs import setup_logger

logger = setup_logger("chatbot", log_filename="warung.log")

logger.info("========================== Chatbot Starting ==============================")

params_produk = {"command": "uv", "args": ["run", "app/mcp_sample/produk_server.py"]}
params_transaction = {"command": "uv", "args": ["run", "app/mcp_sample/transaction_server.py"]}
params_session = {"command": "uv", "args": ["run", "app/mcp_sample/session_server.py"]}

async def warung_ai_chat_with_history(message: str, chat_history: list):
    """
    Version that includes chat history context
    """
    try:
        async with MCPServerStdio(params=params_produk) as mcp_produk:
            async with MCPServerStdio(params=params_transaction) as mcp_transaction:
                
                # Build conversation context from history
                conversation_context = ""
                if chat_history:
                    conversation_context = "\n\nKonteks Percakapan Sebelumnya:\n"
                    for i, (user_msg, ai_response) in enumerate(chat_history):
                        conversation_context += f"User: {user_msg}\nAI: {ai_response}\n\n"
                    conversation_context += "---Percakapan Berlanjut---\n"
                
                # Add context to current message
                full_message = conversation_context + f"User: {message}"
                logger.info(f"full_message: {full_message}")
                
                agent = Agent(
                    name="WarungAI_Agent_Gradio", 
                    instructions=instructions, 
                    model=model, 
                    mcp_servers=[mcp_produk, mcp_transaction]
                )
                
                # trace("percakapan_gradio")
                result = await Runner.run(agent, full_message, max_turns=50) 
                
                print(full_message)
                print("===============================================================\n")
                return result.final_output
                
    except Exception as e:
        print(f"Error dalam warung_ai_chat: {e}")
        return f"Maaf, terjadi kesalahan internal pada server: {str(e)}"


if __name__ == "__main__":
    # Use the first solution
    iface = gr.ChatInterface(
        fn=warung_ai_chat_with_history,  # <-- Use the fixed function
        title="Warung AI Assistant (MCP Demo)",
        description=(
            "Selamat datang di Warung AI! Ketik pesan Anda untuk berinteraksi.\n"
            "AI akan mengingat konteks percakapan sebelumnya.\n"
            "Contoh alur:\n"
            "1. 'Saya mau beli lampu' → AI akan cari produk\n"
            "2. '2' → AI akan proses sebagai kuantitas untuk produk sebelumnya\n"
        ),
        chatbot=gr.Chatbot(height=550, label="Percakapan", show_label=True, show_copy_button=True),
        textbox=gr.Textbox(placeholder="Ketik pesan ke Warung AI...", container=False, scale=7, label="Input Pengguna"),
        theme="soft",
        examples=[
            ["Ada aqua ga?"],
            ["Ada tambahan barang lain?"],
            ["Lanjut pembayaran"],
        ],
        submit_btn=True,
        autofocus=True,
    )
    
    print("Meluncurkan Gradio Interface dengan Context History...")
    iface.launch(show_error=True)
```

```python
import os
import json
from agents.mcp import MCPServerStdio
from agents import Agent, Runner # Pastikan Runner.run mengembalikan objek yang propertinya bisa kita akses
from typing import TypedDict, List, Annotated, Optional # Tambahkan Optional
from langgraph.graph.message import add_messages
from langchain_core.messages import BaseMessage

# ==============================================================================
# 1. DEFINISI STATE (Telah Diperbarui)
# ==============================================================================
class WarungState(TypedDict):
    """
    Mewakili state dari chatbot warung dengan detail transaksi yang kaya.

    Attributes:
        messages: Riwayat percakapan.
        active_transaction: Objek transaksi yang sedang aktif (dictionary), atau None jika tidak ada.
        transaction_details: Daftar objek detail transaksi yang terkait dengan active_transaction.
    """
    messages: Annotated[list, add_messages]
    active_transaction: Optional[dict]  # Bisa None jika tidak ada transaksi aktif
    transaction_details: List[dict]

# ==============================================================================
# 2. INISIALISASI DAN SETUP AWAL
# ==============================================================================
params_produk = {"command": "uv", "args": ["run", "app/mcp_sample/produk_server.py"]}
params_transaction = {"command": "uv", "args": ["run", "app/mcp_sample/transaction_server.py"]}
model = "gpt-4o-mini"

# Ganti dengan nama file instruksi SOP Anda
INSTRUCTION_FILE_PATH = f"{os.getcwd()}/app/instruction_sop.md"
if not os.path.exists(INSTRUCTION_FILE_PATH):
    # Buat file instruksi default jika tidak ada, untuk menghindari error
    print(f"Peringatan: File instruksi '{INSTRUCTION_FILE_PATH}' tidak ditemukan. Menggunakan instruksi default.")
    instructions = "Anda adalah asisten warung. Pandu pengguna untuk membeli produk."
    # Sebaiknya buat file instruction_sop.md dengan instruksi yang detail
else:
    with open(INSTRUCTION_FILE_PATH, 'r', encoding='utf-8') as file:
        instructions = file.read()

# ==============================================================================
# 3. NODE AGENT CERDAS (Implementasi dengan State Transaksi Lengkap)
# ==============================================================================
async def get_response_from_agent(state: WarungState):
    """
    Node agent cerdas yang mengelola state transaksi dan detail transaksi secara lengkap.
    """
    print("\n--- MASUK NODE AGENT (State Transaksi Lengkap) ---")
    current_messages = state['messages']
    active_transaction = state.get('active_transaction')
    transaction_details = state.get('transaction_details', [])

    # === A. BUAT PROMPT YANG KAYA KONTEKS ===
    state_summary_parts = []
    if active_transaction:
        state_summary_parts.append(f"Transaksi Aktif (ID: {active_transaction.get('id', 'N/A')}):")
        state_summary_parts.append(f"  Status: {active_transaction.get('status', 'N/A')}")
        state_summary_parts.append(f"  Total Estimasi Sejauh Ini: Rp {active_transaction.get('total_harga_transaksi', 0):,}")
        if transaction_details:
            state_summary_parts.append("  Detail Item:")
            for i, detail in enumerate(transaction_details):
                state_summary_parts.append(f"    {i+1}. Produk ID: {detail.get('produk_id')}, Qty: {detail.get('qty')}, Subtotal: Rp {detail.get('total_harga_produk', 0):,}")
        else:
            state_summary_parts.append("  Belum ada item di transaksi ini.")
    else:
        state_summary_parts.append("Saat ini belum ada transaksi aktif.")
    
    state_summary = "\n".join(state_summary_parts)
    history_string = "\n".join([f"{msg.type.upper()}: {msg.content}" for msg in current_messages])
    
    full_message_prompt = (
        "== BRIEFING STATUS TRANSAKSI ==\n"
        f"{state_summary}\n"
        "================================\n\n"
        "== RIWAYAT PERCAKAPAN ==\n"
        f"{history_string}\n"
        "========================\n\n"
        "Instruksi: Berdasarkan briefing dan riwayat, tentukan langkah selanjutnya sesuai SOP. Panggil tool yang diperlukan."
    )
    
    print(f"Generated Prompt for LLM:\n{full_message_prompt}")

    # === B. PANGGIL AGENT ===
    async with MCPServerStdio(params=params_produk) as mcp_produk:
        async with MCPServerStdio(params=params_transaction) as mcp_transaction:
            agent = Agent(
                name="WarungAIAgent_SOP_V2",
                instructions=instructions,
                model=model,
                mcp_servers=[mcp_produk, mcp_transaction]
            )
            # `result` adalah objek yang dikembalikan oleh Runner.run
            result = await Runner.run(agent, full_message_prompt, max_turns=50)

    # =======================================================================
    # PENTING: INSPEKSI DAN ADAPTASI ATRIBUT `RunResult`
    # =======================================================================
    print("--- HASIL DEBUG `RunResult` (Periksa ini untuk nama atribut yang benar!) ---")
    print(vars(result)) # <-- JALANKAN INI DAN LIHAT OUTPUTNYA!
    print("--------------------------------------------------------------------------")
    
    # GANTI placeholder_tool_calls_list_attribute_name DENGAN YANG ANDA TEMUKAN DI vars(result)
    placeholder_tool_calls_list_attribute_name = 'tool_calls_placeholder' # <--- GANTI INI (misal: 'tool_calls', 'intermediate_steps')
    
    # GANTI placeholder_tool_name_in_call_attribute_name DENGAN YANG ANDA TEMUKAN
    placeholder_tool_name_in_call_attribute_name = 'tool_name_placeholder' # <--- GANTI INI (misal: 'name', 'tool_name')
    
    # GANTI placeholder_tool_output_in_call_attribute_name DENGAN YANG ANDA TEMUKAN
    placeholder_tool_output_in_call_attribute_name = 'tool_output_placeholder' # <--- GANTI INI (misal: 'output', 'result_str')
    
    # === C. PROSES OUTPUT TOOL UNTUK MEMPERBARUI STATE ===
    updated_active_transaction = active_transaction
    updated_transaction_details = list(transaction_details) # Salin untuk dimodifikasi

    if hasattr(result, placeholder_tool_calls_list_attribute_name):
        tool_calls = getattr(result, placeholder_tool_calls_list_attribute_name, [])
        if isinstance(tool_calls, list):
            print(f"Mendeteksi {len(tool_calls)} pemanggilan tool untuk diproses.")
            for tool_call_info in tool_calls:
                if not isinstance(tool_call_info, dict):
                    print(f"Peringatan: Format tool_call_info tidak terduga: {tool_call_info}")
                    continue

                tool_name = tool_call_info.get(placeholder_tool_name_in_call_attribute_name)
                tool_output_str = tool_call_info.get(placeholder_tool_output_in_call_attribute_name)

                if not tool_name or not tool_output_str:
                    print(f"Peringatan: Informasi tool tidak lengkap di: {tool_call_info}")
                    continue
                
                try:
                    output_data = json.loads(tool_output_str)

                    if tool_name == 'create_transaction' and 'id' in output_data:
                        updated_active_transaction = output_data # Simpan seluruh objek transaksi
                        
                        # PERHATIAN: Baris berikut dikomentari berdasarkan permintaan pengguna ("jangan diapus").
                        # Ini berarti jika transaksi baru dibuat, item dari state 'transaction_details' sebelumnya
                        # TIDAK akan dihapus. Ini bisa menyebabkan item dari sesi belanja yang belum selesai
                        # secara konseptual terbawa ke transaksi formal yang baru.
                        # Efek ini mungkin diinginkan jika state 'active_transaction' sering di-reset
                        # oleh lapisan UI (misalnya Gradio) antar giliran, tetapi 'transaction_details'
                        # (secara hipotetis) tetap ada atau pengguna mengharapkannya tetap ada.
                        # Namun, untuk sistem POS yang bersih, setiap transaksi baru (tool create_transaction)
                        # idealnya dimulai dengan keranjang kosong.
                        # Masalah mendasar kemungkinan adalah bagaimana state (active_transaction & transaction_details)
                        # dipertahankan antar interaksi di level UI (Gradio).
                        # updated_transaction_details = [] # <-- DIKOMENTARI SESUAI PERMINTAAN
                        
                        print(f"STATE_UPDATE: Transaksi baru dibuat. Data: {updated_active_transaction}. Detail transaksi sebelumnya TIDAK dihapus (jika ada).")

                    elif tool_name == 'create_detail_transaction' and 'id' in output_data:
                        if updated_active_transaction and output_data.get('transaction_id') == updated_active_transaction.get('id'):
                            updated_transaction_details.append(output_data)
                            print(f"STATE_UPDATE: Detail transaksi baru ditambahkan: {output_data}")
                        else:
                            print(f"Peringatan: Detail transaksi {output_data.get('id')} tidak cocok dengan transaksi aktif ID {updated_active_transaction.get('id') if updated_active_transaction else 'None'}.")
                    
                    elif tool_name == 'update_transaction' and 'status' in output_data:
                        if output_data['status'] == 'success':
                            if updated_active_transaction and output_data.get('id') == updated_active_transaction.get('id'):
                                updated_active_transaction = output_data 
                                print(f"STATE_UPDATE: Transaksi (ID: {updated_active_transaction.get('id')}) berhasil diupdate & selesai.")
                                # Pertimbangkan untuk mereset active_transaction dan transaction_details di sini
                                # jika AI mengonfirmasi bahwa sesi belanja telah selesai sepenuhnya.
                                # Untuk saat ini, biarkan SOP AI yang memandu kapan transaksi benar-benar baru.
                                # Jika ingin reset otomatis setelah sukses:
                                # updated_active_transaction = None
                                # updated_transaction_details = []
                            else:
                                print(f"Peringatan: Update transaksi {output_data.get('id')} tidak cocok dengan transaksi aktif.")
                        else: 
                             if updated_active_transaction and output_data.get('id') == updated_active_transaction.get('id'):
                                updated_active_transaction = output_data 
                                print(f"STATE_UPDATE: Status transaksi (ID: {updated_active_transaction.get('id')}) diperbarui: {output_data.get('status')}")

                except (json.JSONDecodeError, TypeError) as e:
                    print(f"Info: Gagal memproses output dari tool '{tool_name}'. Mungkin bukan JSON. Output: '{tool_output_str}', Error: {e}")
                    continue
        else:
            print(f"Peringatan: Atribut '{placeholder_tool_calls_list_attribute_name}' bukan list.")
    else:
        print(f"Info: Tidak ada atribut '{placeholder_tool_calls_list_attribute_name}' pada objek `RunResult` atau nilainya kosong.")

    print("--- SELESAI NODE AGENT ---")
    
    return {
        "messages": [("ai", result.final_output)],
        "active_transaction": updated_active_transaction,
        "transaction_details": updated_transaction_details,
    }

# ==============================================================================
# Sisa kode Anda untuk mendefinisikan graph LangGraph dan menjalankan Gradio
# ==============================================================================
```

```python
from langgraph.graph import StateGraph, END

# Inisialisasi graph
workflow = StateGraph(WarungState)

# Tambahkan nodes ke dalam graph
workflow.add_node("agent", get_response_from_agent)
# workflow.add_node("user_input", handle_user_input) # Contoh node lain

# Tentukan entry point
workflow.set_entry_point("agent")
workflow.add_edge("agent", END)

app = workflow.compile()
```

```python
import gradio as gr


async def chat_interface_fn(message: str, history: list):
    """Fungsi yang akan dijalankan oleh Gradio Interface."""

    messages = []
    for user_msg, ai_msg in history:
        messages.append(("user", user_msg))
        messages.append(("ai", ai_msg))
    messages.append(("user", message))

    initial_state = {
        "messages": messages,
        "transaction_id": None, 
        "cart_items": []      
    }


    final_state = await app.ainvoke(initial_state)
    response = final_state['messages'][-1].content

    logger.info(f"full message: \nuser intput: ${message} \nAI response: ${response}")
    
    return response

if __name__ == "__main__":
    iface = gr.ChatInterface(
        fn=chat_interface_fn,
        title="Warung AI Assistant (LangGraph Version)",
        description="Chatbot dengan state dan memory menggunakan LangGraph.",
        theme="soft",
    )
    iface.launch(show_error=True)
```
